#### Create new features: next commit, status of issue being open or closed depending on the existence of a particular error in the next commits

```python
def lag_commit_features(dataframe, repository):
    dataframe = dataframe.loc[dataframe['repository'] == repository].copy()
    # identify the first value of each stretch
    m = dataframe['commit_hash'].ne(dataframe['commit_hash'].shift())
    # mask the other, shift up, backfill
    dataframe['next_commit_hash'] = dataframe['commit_hash'].where(m).shift(-1).bfill()
    
    dataframe = status_feature(dataframe)

    return dataframe

def status_feature(df):
    df['status'] = ""

    for commit in df['commit_hash'].unique():
        print("Checking commit hash: {}".format(commit))
        df_com = df[df['commit_hash'] == commit]
        next_commit = df[df['commit_hash'] == df_com['next_commit_hash'].iloc[0]]

        common_messages = np.intersect1d(df_com['message'], next_commit['message'])

        df.loc[df['commit_hash'] == commit, "status"] = "closed"
        df.loc[(df['commit_hash'] == commit) & df['message'].isin(common_messages), "status"] = "open"

    return df

unique_repo = df_v2['repository'].unique()
temp_list = []
for index,repo in enumerate(unique_repo):
    print("Working for the repository:{} // number:{} // out of:{}".format(repo, index, len(unique_repo)))
    df_temp = lag_commit_features(df_v2, repo)
    temp_list.append(df_temp)

df_pre = pd.concat(temp_list, ignore_index=True)
```

#### Creating feature: error_count depending on the number of the issue in each commit

```python
def error_count(repository, dataframe):
    count = 0 
    msgstack = []
    
    dataframe = dataframe[dataframe['repository'] == repository]
    
    for index,row in dataframe.iterrows():
        if row.message in msgstack:
            msgstack.remove(row.message)
            count = count - 1
            dataframe.loc[index, 'error_count'] = count
            
        else:
            msgstack.append(row.message)
            count = count + 1
            dataframe.loc[index, 'error_count'] = count
        
    return dataframe

df_list = []
for index, repo in enumerate(unique_repo):
    print("Working for the repository:{} // number:{} // out of:{}".format(repo, index, len(unique_repo)))
    temp = error_count(repo, df_pre)
    df_list.append(temp)

# Concatenate the list of DataFrames into a single DataFrame
df_pre_v2 = pd.concat(df_list, ignore_index=True)
```

#### Creating feature: time lapse depending on the existence of issue in particular commit

```python
def time_lapse(repository, dataset):
    dataset['time_lapse'] = ""
    dataset = dataset[dataset['repository'] == repository]
    dataset['committed_date'] = pd.to_datetime(dataset['committed_date'])
    dataset.sort_values(by=['committed_date'])
    start_date = dataset['committed_date'].min()
    dataset['time_lapse'] = dataset['committed_date'].apply(lambda x : x - start_date)
                   
    return dataset
          
 
temp_list = []
for index, repo in enumerate(unique_repo):
    print("Working for the repository:{} // number:{} // out of:{}".format(repo, index, len(unique_repo)))
    temp = time_lapse(repo, df_pre_v2)
    temp_list.append(temp)

# Concatenate the list of DataFrames into a single DataFrame
df_pre_v3= pd.concat(temp_list, ignore_index=True)
```

#### Creating feature: workflow_unique_id depending on the 

```python
def generate_workflow_id(dataset):
    
    dataset = dataset.sort_values(by='committed_date')
    mapping = {}
    
    result = {}
    
    for index, row in dataset.iterrows():
        repo = row['repository']
        filepath = row['file_path']
        hash_val = row['commit_hash']
        previous = row['previous_file_path']
        author = row['author_name']
#         print("The index is :{} and previous:{} | and datatype of previous:{}".format(index, previous, type(previous)))
        
        if filepath and type(previous) == float :
            unique_id = f"{author}/{repo}/{filepath}/{hash_val}"
#             if (repo, filepath) in mapping:
#                 raise ValueError(f"Duplicate file detected: {repo}/{filepath}")
#             print("Unique ID:{}".format(unique_id))
            mapping[(repo, filepath)] = unique_id
            result[index] = unique_id

#         Case 2: A file is removed
        elif not filepath and previous:
            if (repo, previous) not in mapping:
                raise ValueError(f"Cannot find unique ID for removed file: {repo}/{previous}")
            del mapping[(repo, previous)]
            result[index] = None

        # Case 3: A file is renamed
        elif filepath and previous:
            #Some files modified
            unique_id = f"{author}/{repo}/{filepath}/{hash_val}"
            if (repo, previous) not in mapping:
                print("Cannot find unique ID for previous filepath: {repo}/{previous}")
                result[index] = unique_id
                continue

            mapping[(repo, filepath)] = mapping[(repo, previous)]
            result[index] = mapping[(repo, filepath)]
            
            
    return pd.Series(result)
            
            
    
```

#### Resampling the data based on the dates

```python
#helper function
def get_the_missing_dates(date, number_of_days):
    list_date = []
    for i in range (1, number_of_days): 
        date_x = date + timedelta(days= i)
        list_date.append(date_x)
    
    return list_date

def set_the_duplicate_row(number_of_dates,dataset,index, error_sum):
#     print("number of dates:{} and dataset:{} and index:{}".format(number_of_dates,dataset,index))
    new_df = pd.DataFrame()
    new_list = []
    print("length of the dataset:{}".format(len(dataset)))
    print("Number of the rows added:{}".format(len(number_of_dates)))
    for i in number_of_dates:
        row_duplicate = dataset[dataset.index == index]
        row_duplicate['committed_date'] = i
        row_duplicate['date'] = i
        row_duplicate['error_count'] = error_sum
        
        new_list.append(row_duplicate)
        
    
    return new_list
        
        
```

```python
from datetime import timedelta
def resampling(dataset, repository):
    dataset = dataset[dataset['repository'] == repository]
    if len(dataset) < 2:
        return dataset
    else:

        error_count_sum = dataset.groupby('date')[['error_count']].sum()
        dataset = dataset.sort_values(by='committed_date')
        temp_list = []
        dataset_c = dataset.copy()

        for index, row in dataset.iterrows():
            date = row['date']
            next_date = row['next_date']
            number_of_days = (next_date - date).days
            expected_next_date = get_the_missing_dates(date,number_of_days)

            error_sum = error_count_sum[error_count_sum.index == date].values[0][0]
    #         print("The date of index :{} is: {} and the next date is: {} and the differentiation of date is: {} and the expected date is:{} ".format(index,date,next_date, number_of_days, expected_next_date))
            if date == next_date:
                continue

            elif date != next_date and number_of_days > 1:
                data = set_the_duplicate_row(expected_next_date, dataset_c, index, error_sum)
                temp_list.extend(data)

                print("Row inserted !!!")

            else:
                continue


        if not temp_list:      
            return dataset

        else: 
            df_data = pd.concat(temp_list, ignore_index=True)
            combined_df = pd.concat([dataset_c, df_data], ignore_index=True)
            combined_df = combined_df.sort_values(by='committed_date')

            return combined_df
        
```
